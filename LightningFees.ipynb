```python
import os
import pandas as pd
import numpy as np
import datetime

import plotly.graph_objects as go
from plotly.subplots import make_subplots

%load_ext autoreload
%autoreload 2
%config InlineBackend.figure_format = 'retina'

import chart_utils
import analysis_utils
```

```python
fees = pd.read_csv('./data/lightning_fees.csv')
```

```python
fees = fees[:-2]
```

```python
fees.loc[fees.lightning_fees >0].date.min()
```

```python
for col in ['total_fees', 'channel_open_fees', 'channel_close_fees', 'lightning_fees']:
    fees['{}_28d'.format(col)] = fees[col].rolling(28).mean()

fees.columns

fees['lightning_pct_28d'] = fees['lightning_fees_28d'] / fees['total_fees_28d']
```

```python
chart_utils.single_axis_chart2(
    fees.loc[fees.date >= '2018-01-01'], x_series='date', 
    y_series='lightning_pct_28d',
    x_axis_title='Date',
    y_series_axis_format= ',.2%',
    y_series_title='Lightning Public Channel Open + Close Fees (% total)',
    title='Lightning Public Channel Open + Close Fees: % of total Bitcoin blockspace fees',
    data_source="ln.bigsun.xyz",
    halving_lines=False,
)
```
